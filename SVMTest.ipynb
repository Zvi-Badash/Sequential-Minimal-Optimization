```python
from datasetGenerator import generate_linear_separable_dataframe, plot_line
from SupportVectorMachine import SVM
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = (10, 6)
sns.set_style('darkgrid')
sns.set_palette(sns.color_palette('deep'))
```

# Generating a dataset

```python
N, d = 50, 2
target_weights, f, df = generate_linear_separable_dataframe((N, d), seed=34, noise=0.)
X, y = np.array(df.drop(columns='class')), np.array(df['class'])
```

```python
sns.scatterplot(data=df, x='x1', y='x2', hue='class')
plot_line(target_weights, color='k')
```

# Defining the classifier

```python
svc = SVM(C=1, tol=1e-5, seed=3)
svc.fit(X, y)
```
